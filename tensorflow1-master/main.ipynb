```python
import tensorflow as tf
import pandas as pd
```

```python
data = pd.read_csv(r'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv')

독립 = data[['온도']]
종속 = data[['판매량']]

X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)

model = tf.keras.models.Model(X, Y)
model.compile(loss = 'mse')
model.fit(독립.values, 종속.values, epochs = 10000)
```
